## Data Cleaning

```python
# Converting duration from milliseconds to seconds
tracks['duration_s'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop('duration_ms', inplace= True, axis=1)
```

```python
# Converting duration to minutes
tracks['duration_min'] = tracks['duration_s'].apply(lambda x: round(x/60,2))
tracks.drop('duration_s', inplace=True, axis = 1)
```

```python
# Converting data type for Explicit
tracks['explicit'] = tracks['explicit'].astype(bool)
```

```python
# Cleaning artist names
tracks['artists'] = tracks['artists'].str.strip("[]").str.replace("'","")
tracks['artists'][:10]
```

```python
# Changing datatype of release date
tracks['release_date'] = pd.to_datetime(tracks['release_date'], format = 'mixed')
tracks['release_date'][:10]
```

## Data Analysis

```python
# Plotting correlation 
tracks_subset = tracks.drop(['key','mode','explicit','id','artists','id_artists','name'], axis=1)

# Converting the dataframe to numeric
tracks_subset = tracks_subset.apply(pd.to_numeric, errors = 'coerce')

# Calculate correlation
correl = tracks_subset.corr(method='pearson')

# Plotting correlation  
plt.figure(figsize=(20,10))
heatmap = sns.heatmap(correl, annot=True, fmt=".1g",vmin=-1, vmax=1, center=0, cmap='inferno',linewidths=1, linecolor='Black',
                      annot_kws={'size':15,'weight':'bold'})
heatmap = plt.title("Correlation Heatmap between variables")
```

 - Loudness and energy are strongly positively correlated.
 - Acousticness and energy strongly negatively correlated.
 - Release date and popularity are mildly positively correlated.

```python
# Popular Songs
popular_tracks = tracks[tracks['popularity'] >= 70]
popular_tracks.shape
```

```python
# Popular songs by artist
popular_tracks_artist = popular_tracks['artists'].value_counts()    
popular_tracks_artist = popular_tracks_artist.reset_index()
popular_tracks_artist.columns = ['artist','count of popular songs']

popular_tracks_artist.head()
```

```python
# Top 20 Artists
popular_artists = popular_tracks_artist.sort_values(by = 'count of popular songs',ascending=False).head(20)

popular_artists
```

```python
# Plotting Bar Chart
fig, ax = plt.subplots(figsize = (8,5))
ax = sns.barplot(x='artist',y='count of popular songs',data=popular_artists, palette='Spectral',
                 hue = 'count of popular songs',legend=False)
plt.title('Top 20 Popular Artists', c = 'black', fontsize=15)
plt.xlabel('Artist')
plt.xticks(rotation = 80) # Rotating values on x-axis
plt.ylabel('Number of Popular Songs')
plt.show()
```

Taylor Swift is the most popular artist

```python
# Top 10 Songs
top_songs = tracks.sort_values(by='popularity', ascending = False).head(10)
top_songs
```

```python
# Plotting Bar Chart
fig, ax = plt.subplots(figsize = (8,5))
ax = sns.barplot(y='name', x='popularity', data=top_songs,palette='Spectral',hue='popularity', 
                 legend = False)
plt.xlabel('Popularity', fontsize = 12)
plt.ylabel('Song Name', fontsize = 12)
plt.title('Top 10 Popular Songs',fontsize = 15)
plt.show()
```

Peaches is the most popular song

```python
# Plotting Energy, Dancebility, Liveness, Valence, Acousticness and Instrumentalness of these songs

columns_to_plot = ['energy','danceability','liveness','valence','acousticness','instrumentalness']

for i, column in enumerate(columns_to_plot):
    fig,ax = plt.subplots(figsize=(8,5))
    ax = sns.barplot(x='name',y=column, data=top_songs, palette='Spectral',hue=column, legend=False)
    plt.title(f'{column} of Top 10 Songs', fontsize = 15)
    plt.xlabel('Song name', fontsize = 12)
    plt.xticks(rotation=80)
    plt.ylabel(column, fontsize = 12)
    plt.show()
```

```python
# Correlation between Energy and Loudness
plt.figure(figsize = (8,5))
sns.regplot(x='energy',y='loudness',data=tracks_subset, color='lightblue',line_kws={'color':'red'},
            scatter_kws={'s':10}).set(title="Energy vs Loudness Correlation")
plt.grid()
plt.show()
```

Strong positive correlation between Energy and Loudness

```python
# Songs releaed per year
tracks['year'] = tracks['release_date'].dt.year

songs_per_year = tracks['year'].value_counts().sort_index()

plt.figure(figsize=(10,4))
sns.lineplot(x=songs_per_year.index, y=songs_per_year.values, marker='o',color='royalblue',linewidth=2,markersize=5)
plt.title('Songs released By year', fontsize = 15)
plt.ylabel('Count Of Songs', fontsize=10)
plt.xlabel('Year', fontsize = 10)
plt.grid()
plt.show()
```

 - Songs have been released at a constant growth since 1920.
 - Experienced a slight dip in the end of 60's and mid 90's with a signficant dip in Y2K.
 - Number of Songs released saw an all-time high in 2020
